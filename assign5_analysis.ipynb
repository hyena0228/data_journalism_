```python
#tagged_sentences = [tagger.pos(tweet) for tweet in tweet_list]
tokens = [nltk.word_tokenize(tweet) for tweet in tweet_list]
tokens

tokens = sum(tokens, [])
print(tokens)
```

```python
tagged = nltk.pos_tag(tokens)
tagged
```

```python
import nltk
nltk.download('omw-1.4')

lemma = nltk.wordnet.WordNetLemmatizer()
gutenberg_lemma = []
for token in tokens:
    gutenberg_lemma.append(lemma.lemmatize(token))

gutenberg_lemma
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words
```

```python
stop_words.update(['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}', 'rt', '#', '@'])
filtered_words = [word.lower() for word in gutenberg_lemma if word.lower() not in stop_words]
filtered_words
```

```python
from collections import Counter
word_counts = Counter(filtered_words)
word_counts.most_common()
```
