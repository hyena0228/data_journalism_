## 1. 미디어다음에서 제공하는 영어사전 웹사이트에서 랜돔으로 100개의 단어를 수집하고, 단어와 단어의 의미를 보기 좋게 출력하시오. (5 Points)

예: http://dic.daum.net/word/view.do?wordid=ekw000078303&q=help


* 저장 형식은 자유. 단, 가능한 한 보기 좋게 내용을 편집하여 저장하자.

  예)
  
    단어 1 
      1. 뜻풀이  
      2. 뜻풀이 
    ===  
    단어 2 
      1. 뜻풀이

```python
import random

url_list = []

for i in range(100):
    num = random.randrange(1,300000)
    url_list.append("https://dic.daum.net/word/view.do?wordid=ekw{0:09d}".format(num))
    
url_list
```

```python
import urllib.request
from bs4 import BeautifulSoup
import time


for i in range(100):
    with urllib.request.urlopen(url_list[i]) as url:
        doc = url.read()
        soup = BeautifulSoup(doc, "html.parser")
        word = soup.find_all("span", class_ = 'txt_cleanword')[0].text.strip()
        means_list = soup.find_all("span", class_ = 'txt_mean')
        print(i+1, end = ' ')
        print(word)
        #print(means_list)
        num = 1
        for mean in means_list:
            print(' ', end = '')
            print(num, end = ' ')
            print(mean.text)
            num += 1
            #print(mean)
        print('=================')
        time.sleep(0.1)
```
