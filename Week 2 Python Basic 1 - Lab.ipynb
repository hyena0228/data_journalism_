## 4.2 출력 옵션

여기서 
```
주문하신 음료는 
아메리카노
입니다.
```
와 같이 표시된 것이 별로 보기 좋지 않다. 세 줄의 문자열을 한줄로 만들어주고 싶다.

이런 문제를 해결하기 위해서는 출력 `print()`의 옵션을 변경해주면 다.

`print()`는 기본적으로 출력할 때 마지막에 줄바꿈을 하도록 설정되어 있다. 

이것을 줄바꿈이 아닌 띄어쓰기로 바꾸고 싶다면 `print()` 안에 `,end=' '` 옵션을 넣어준다.

````python
menu = input('무엇을 주문하시겠습니까?\n')
print('주문하신 음료는', end=' ')
print(menu, end=' ')
print('입니다.')
````

아메리카노 입니다를 아메리카노입니다로 붙여쓰고 싶다면? 띄어쓰기를 없애준다. 

`'  '` 대신 `''` 을 입력하면 된다.

````python
menu = input('무엇을 주문하시겠습니까?\n')
print('주문하신 음료는', end=' ')
print(menu, end='')
print('입니다.')
````

## 6.1 조건문 (if)

카드인지 현금인지에 따라 다른 결과를 보여줄 때는 조건문(conditional statement)을 사용한다. 조건문은 프로그램에서 '분기'를 만들어준다. 분기란 주어진 조건에 따라 다른 결과를 보여주는 것을 얘기한다.

가장 대표적인 `if statement`를 이용하여 조건문을 만들어 보자.

`if statement`은 **조건**과 **실행문**이 필요하다.

````python
if (조건) :
  실행
````

조건문은 **조건(condition)**은 수학의 논리 관계를 이용하여 참과 거짓이 명확해야 한다. 
```
1. a랑 b가 같다: a == b
2. a랑 b가 같지 않다: a != b
3. a가 b보다 크다: a > b
4. a가 b보다 작다: a < b
5. a가 b보다 크거나 같다: a >= b
6. a가 b보다 작거나 같다: a <= b
```

if 문의 예시를 하나 들어보자.


````python
amount = 4
if amount > 2:
  print("a) 캐리어가 필요하신가요?")

amount = 2
if amount > 2:
  print("b) 캐리어가 필요하신가요?")
````

## 7.1 주석 (Comment)

코드가 길어지면 한 눈에 알아보기가 어렵다.

이럴 때 주석을 사용한다.

주석은 코드에서 **실행이 안되는 코드**로 보통 설명을 할 때 사용한다.

````python
# 이 프로그램은 절대 실행되지 않는다
print("실행 되는데?")
````

위의 코드에 주석을 넣어 프로그램의 가독성을 높여보자.
